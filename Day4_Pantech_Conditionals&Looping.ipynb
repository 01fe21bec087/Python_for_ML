# ***Conditional Statements***

## ***if statement***

```python
a=3
b=5
```

```python
# if
if a<b:
  print("b is greater than a")
```

## ***if-else statement***

```python
# if-else
if a>b:
  print("Hi")
else:
  print("Bye")
```

## ***elif statement***

```python
# elif
if a==b:
  print("Equal")
elif a<b:
  print("Less")
else:
  print("More")
```

```python
db_user="John"
db_pass="abcd@123"

user=input("Enter username: ")
password=input("Enter password: ")

if (user==db_user) and (password==db_pass):
  print("Login successful")
else:
  print("Login failed")
```

## ***nested if statement***

```python
#Nested if statement
x=25

if x>10:
  print("Above ten,")
  if x>20:
    print("and also above 20!")
  else:
    print("but not above 20.")
else:
  print("Below ten")
```

## ***pass statement***

```python
# Pass statement
# to do nothing in if body
if x<10:
  pass
```

# ***Looping statement (Iteration)***

## ***for loop***

Used to iterate over a sequence (list, tuple, string)
Strings are iterable objects: they contain sequence of characters

```python
for x in "banana":
  print(x)
  print(type(x))
```

```python
fruits=["apple","banana","cherry","kiwi"]
for x in fruits:
  print(x)
```

```python
for x in fruits:
  for y in x:
    print(y)
  print()
```

```python
# Finding string length without len()
name="Abhinav"
l=0
for n in name:
  l+=1
print(l)
```

### ***Break statement***

```python
print(fruits)
```

```python
for x in fruits:
  print(x)
  if(x=="cherry"):
    print("We found",x)
    break
```

### ***continue statement***

```python
for x in fruits:
  if x=="banana":
    print("We are skipping",x)
    continue
  print(x)
```

### ***Range***

```python
# list of integers from 0 to <10
list(range(10))
```

```python
# list of integers from 4 to <10
list(range(4,10))
```

### ***Nested for loop***

```python
adj=["red","big","tasty"]
fruit=["apple","banana","cherry"]
for x in adj:
  for y in fruit:
    print(x,y)
```

## ***While loop***

While loop keeps on running till the condition is not false

i.e. while loop executes if condition is not false

```python
while True:
  print("Hello")
```

```python
x=1
while x<10:
  if x==7:
    break
  print(x)
  x+=1
```

### ***while-else statement***

```python
x=1
while x<10:
  print(x)
  x+=1
else:
  print("x is no longer less than 10")
```

```python
x=1
while x<10:
  if x==6:
    break
  print(x)
  x+=1
else:
  print("x is no longer less than 10")
```
