## ***Introduction to Matplotlib***

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,1,10)
m=0.6
c=25
y=m*x+c
plt.plot(x,y)
plt.show()
```

## ***Scatter Plots***

```python
#temperature dependance of resistance for thermistor
import matplotlib.pyplot as plt
import numpy as np

Temp=np.array([10,20,25,30,40,50,60,70,80,90,100,110,120])
Res=np.array([18476,12185,10000,8260,5740,4080,2960,2188,1645,1257,973,765,608])

#Plot1
plt.scatter(Temp,Res)
plt.show()

#Plot2
fig, a=plt.subplots()
a.plot(Temp,Res,linewidth=2.0)
plt.show()

#Plot3
fig, a=plt.subplots()
a.plot(Temp,Res,linewidth=1.0)
plt.scatter(Temp,Res)
#plt.show()
```

## ***Garnishing Graphs***

```python
#Garnishing graphs
import matplotlib.pyplot as plt
import numpy as np
Temp=np.array([10,20,25,30,40,50,60,70,80,90,100,110,120])
Res=np.array([18476,12185,10000,8260,5740,4080,2960,2188,1645,1257,973,765,608])

plt.plot(Temp,Res)
plt.show()

plt.plot(Res,marker='*')
plt.show()

plt.plot(Res,linestyle='dotted')
plt.show()
```

```python
#Using grid and labels
import matplotlib.pyplot as plt
import numpy as np
Temp=np.array([10,20,25,30,40,50,60,70,80,90,100,110,120])
Res=np.array([18476,12185,10000,8260,5740,4080,2960,2188,1645,1257,973,765,608])

plt.title("Thermistor Characteristics")
plt.xlabel("Temperture")
plt.ylabel("Resistance")

plt.plot(Temp,Res)
plt.grid()
plt.show()
```

## ***Bar Graphs and Pie chart***

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array(["Karnataka","Andhra.P","Tamil.N","Telangana"])
y=np.array([65,68,89,86])

#Bar graph
plt.bar(x,y)
plt.show()

#Pie chart
plt.pie(y)
plt.show()
```
